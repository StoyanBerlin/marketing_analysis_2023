```python
def replace_values(x):
    if x == 'Alone':
        return 'Single'
    elif x == 'YOLO':
        return 'Together'
    elif x == 'Absurd':
        return 'Single'
    else:
        return x

df_clean['Marital_Status'] = df_clean['Marital_Status'].apply(replace_values)
```
